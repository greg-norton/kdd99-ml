Looks like the Kaggle dataset was labeled "Test" and "Train", but they didn't drop the labels from the "Test" data. When compared to the "kddcup.testdata.unlabeled_10_percent" data, there are two coloumns, namely 42 and 43 that have been dropped from the unlabeled data

column 42 is "attack" -- indicates which of the attacks the row of data represents from:

**back,buffer_overflow,ftp_write,guess_passwd,imap,ipsweep,land,loadmodule,multihop,neptune,nmap,normal,perl,phf,pod,portsweep,rootkit,satan,smurf,spy,teardrop,warezclient,warezmaster**.

**normal** represents that the traffic was not categorized as an attack.

column 43 is "last_flag". I'm not sure what this is, but it seems to be considered a label. (not in the unlabeled data)

Since I don't know what last_flag is, I am going to drop it for now and train on "attack"

Next, we need to Convert **protocol_type, service, flag, and attack** columns which is are objects in the dataframe to a discrete numerical value.

I'm following along the Pandas in TensorFlow guide here:
https://www.tensorflow.org/tutorials/load_data/pandas_dataframe


```python
# this encodes udp as 2, tcp as 1, etc. 
df['protocol_type'] = pd.Categorical(df['protocol_type'])
print(df.protocol_type.cat.codes)
df['protocol_type'] = df.protocol_type.cat.codes
```
